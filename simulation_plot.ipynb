```python
import pandas as pd

# Load the uploaded CSV file to check its structure
file_path = '2025-01-04_Namespace(component_size=10, num_paterns=5, trials=5, epochs=1500, lr=0.01, early_stopping=10, split_ratio=0.5, num_stepdata=40, all_path=False, Regularization=\'L1\', hyperparameter=0.001).csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand the structure of the data
data.head()
```

```python
import matplotlib.pyplot as plt

# # Select the first and fourth columns ('KL_提案手法' and 'KL_比較手法')
# column_data_1 = data['KL_提案手法']
# column_data_4 = data['KL_比較手法']

# # Split data into 5 groups of 5 rows each and calculate means for both columns
# grouped_means_1 = [column_data_1[i:i+5].mean() for i in range(0, len(column_data_1), 5)]
# grouped_means_4 = [column_data_4[i:i+5].mean() for i in range(0, len(column_data_4), 5)]
# Prepare grouped means for all columns from 1 to 6

grouped_means = {}
for idx, col in enumerate(data.columns[1:7], start=0):  # Enumerate with numeric keys starting from 0
    grouped_means[idx] = [data[col][i:i+5].mean() for i in range(0, len(data[col]), 5)]

```

```python
# Plot multiple boxplots with different y-axis labels and grouped means
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=False)  # Create 3 subplots

# Data and labels for each subplot
data_groups = [
    (grouped_means[0], grouped_means[3], "KL divergence"),
    (grouped_means[1], grouped_means[4], "Hellinger distance"),
    (grouped_means[2], grouped_means[5], "JS divergence"),
]

colors = ["lightblue", "lightgreen"]
titles = ["Proposed method", "Comparison method"]

# Loop through each subplot and plot
for ax, (data1, data2, ylabel) in zip(axes, data_groups):
    box = ax.boxplot(
        [data1, data2],
        vert=True,
        widths=0.5,
        patch_artist=True,
        tick_labels=titles
    )

    # Customize colors for each box
    for patch, color in zip(box['boxes'], colors):
        patch.set_facecolor(color)

    # Set title and labels
    # ax.set_title("10 KCs", fontsize=14)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
    ax.tick_params(axis='y', labelsize=10)
    # ax.set_ylim(0, None) # 0まで表示

    fig.legend(
    labels=["Proposed method", "Comparison method"],
    loc="upper center",
    ncol=4,
    fontsize=12,
    bbox_to_anchor=(0.5, 1.1)
    )

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

```
